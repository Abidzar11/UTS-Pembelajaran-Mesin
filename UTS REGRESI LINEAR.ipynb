```python
# Linear Regression on Rice Field Dataset

## Importing Libraries
import pandas as pd  # Untuk manipulasi dan analisis data
import numpy as np  # Untuk operasi numerik
from sklearn.model_selection import train_test_split  # Untuk membagi data menjadi set latih dan uji
from sklearn.linear_model import LinearRegression  # Model regresi linear
from sklearn.metrics import mean_squared_error, r2_score  # Untuk evaluasi model
import matplotlib.pyplot as plt  # Untuk visualisasi
import seaborn as sns  # Untuk visualisasi yang lebih menarik
```

```python
## Load Dataset
file_path = 'R02_rice_field.csv'  # file dataset
rice_data = pd.read_csv(file_path)  # Membaca dataset dari file CSV
```

```python
## Display Dataset Information
print("Dataset Info:")  # Menampilkan informasi dataset seperti tipe data dan jumlah nilai kosong
rice_data.info()
```

```python
print("\nFirst 5 rows of the dataset:")  # Menampilkan 5 baris pertama dataset
print(rice_data.head())
```

```python
## Data Description
print("\nDataset Description:")  # Menampilkan statistik deskriptif dari dataset
print(rice_data.describe())
```

```python
## Split Data into Training and Testing Sets
X = rice_data[['Production (X)']]  # Fitur input (Produksi padi)
y = rice_data['Harvest Area (Y)']  # Target output (Luas lahan panen)
```

```python
# Membagi dataset menjadi data latih (70%) dan data uji (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
## Train the Linear Regression Model
model = LinearRegression()  # Inisialisasi model regresi linear
model.fit(X_train, y_train)  # Melatih model dengan data latih
```

```python
## Make Predictions
y_pred = model.predict(X_test)  # Membuat prediksi pada data uji
```

```python
## Evaluate the Model
# Menghitung Mean Squared Error (MSE) dan R-squared (R²) untuk evaluasi performa model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("\nMean Squared Error (MSE):", mse)  # MSE: Rata-rata kesalahan kuadrat
print("R-squared (R2) Score:", r2)  # R²: Seberapa baik model menjelaskan variabilitas data
```

```python
## Visualize the Model
plt.figure(figsize=(10, 6))  # Mengatur ukuran plot
sns.scatterplot(x=y_test, y=y_pred, color='blue', label='Predicted vs Actual')  # Scatter plot untuk membandingkan nilai aktual dan prediksi
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Ideal Fit')  # Garis ideal
plt.xlabel('Actual Harvest Area')  # Label sumbu X
plt.ylabel('Predicted Harvest Area')  # Label sumbu Y
plt.title('Actual vs Predicted Harvest Area')  # Judul plot
plt.legend()  # Menampilkan legenda
plt.show()  # Menampilkan plot
```
